# RainFall

```python
zinkwan = sorted(glob("data/rf/남양주시(진관교)/*.csv"))
daegok = sorted(glob("data/rf/서울시(대곡교)/*.csv"))
songjeong = sorted(glob("data/rf/서울시(송정동)/*.csv"))
```

```python
loc_name=['zinkwan','daegok','songjeong']
rf=[zinkwan,daegok,songjeong]

for idx,i in enumerate(loc_name):
    rf_df=[]
    for file in rf[idx]:
        tmp = pd.read_csv(file)
        rf_df.append(tmp)
    globals()['{}_df'.format(i)] = pd.concat(rf_df, axis=0)
    globals()['{}_df'.format(i)].reset_index(drop=True,inplace=True)
    globals()['{}_df'.format(i)].drop(['rfobscd','links'],axis=1,inplace=True)
```

```python
zinkwan_df['ymdhm'] = pd.to_datetime(zinkwan_df['ymdhm'], format='%Y%m%d%H%M')
daegok_df['ymdhm'] = pd.to_datetime(daegok_df['ymdhm'], format='%Y%m%d%H%M')
songjeong_df['ymdhm'] = pd.to_datetime(songjeong_df['ymdhm'], format='%Y%m%d%H%M')
```

```python
zinkwan_df.columns=['ymdhm','rf_zinkwan']
daegok_df.columns=['ymdhm','rf_daegok']
songjeong_df.columns=['ymdhm','rf_songjeong']
```

```python
rf_df=[zinkwan_df,daegok_df,songjeong_df]
rf = reduce(lambda left, right: pd.merge(left, right, on='ymdhm'), rf_df)
rf.replace(" ",np.nan,inplace=True)
```

```python
rf = rf.sort_values(by='ymdhm', ascending=True)
rf.reset_index(inplace=True,drop=True)
```

```python
rf
```

```python
rf.to_csv('data/rf/origin_rf.csv', index = None)
```

# Tide_lv

```python
tide_lv = sorted(glob("data/tide_level/강화대교/*.csv"))
```

```python
tide_list=[]
for file in tide_lv:
    tmp = pd.read_csv(file)
    tide_list.append(tmp)
tide_df = pd.concat(tide_list, axis=0)
tide_df.reset_index(drop=True,inplace=True)
#tide_df.drop(['dmobscd','links'],axis=1,inplace=True)
tide_df['record_time'] = pd.to_datetime(tide_df['record_time'], format='%Y-%m-%d %H:%M:%S')
tide_df=tide_df.sort_values(by='record_time')
tide_df=tide_df.reset_index(drop=True)
```

```python
tide_df.columns=['ymdhm','tide_level']
```

```python
tide_df.replace(" ",np.nan,inplace=True)
tide_df = tide_df.sort_values(by='ymdhm', ascending=True)
tide_df.reset_index(inplace=True,drop=True)
tide_df = tide_df.drop_duplicates(subset='ymdhm', keep='first')
tide_df.to_csv('data/tide_level/origin_tidelv.csv', index = None)
```
